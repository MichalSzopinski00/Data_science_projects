```python

%matplotlib inline
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.style.use('ggplot')
```

```python
new_df = pd.read_csv("https://cocl.us/datascience_survey_data")
print("import data completed")
```

```python
new_df = new_df.set_index('Unnamed: 0')
```

```python
new_df.index.rename(' ', inplace=True)
```

```python
new_df
```

```python
#sorted by very interested descending
df_new1 = new_df.sort_values(['Very interested'],ascending=False)
df_new1
```

```python
#round 
i=0
x=0
siema=round(df_new1.iloc[i]/2233,2)
while(i<=5):
    df_new1.iloc[x] = round(df_new1.iloc[i]/2233,2)
    i+=1
    x+=1
```

```python
ax=df_new1.plot(kind="bar",
             figsize=(20,8),
            color=['#5cb85c','#5bc0de','#d9534f'],
            width=0.8,
            fontsize=14)
ax.set_title('Percentage of Respondents Interests in Data Science Areas')
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
    
plt.show()
```

```python
df_2 = pd.read_csv('https://cocl.us/sanfran_crime_dataset')
```

```python
# clean up the dataset to remove unnecessary columns (eg. REG) 
df_2 = df_2.groupby(['PdDistrict']).count().reset_index()
df_2.drop(['Category','DayOfWeek','Date','Time','Resolution','Location','Address','X','Y','Descript','PdId'], axis = 1, inplace = True)

# let's rename the columns so that they make sense
df_2.rename(columns={'PdDistrict':'Neighborhood', 'IncidntNum':'Count'}, inplace=True)

# for sake of consistency, let's also make all column labels of type string
df_2.columns = list(map(str, df_2.columns))


```
